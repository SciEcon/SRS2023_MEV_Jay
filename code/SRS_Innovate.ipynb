```python
from web3 import Web3
import pandas as pd


# Connect to the Ethereum mainnet using an Infura API endpoint
web3 = Web3(Web3.HTTPProvider('https://mainnet.infura.io/v3/YOUR_API_KEY'))

# block numbers
block_numbers = [10000000, 10000001, 10000002]

# export transactions from a block
def export_transactions(block_number):
    block = web3.eth.getBlock(block_number)
    transactions = block['transactions']

    transaction_list = []
    for tx_hash in transactions:
        tx = web3.eth.getTransaction(tx_hash.hex())
        receipt = web3.eth.getTransactionReceipt(tx_hash.hex())

        transaction_list.append({
            'Block': block_number,
            'Timestamp': pd.to_datetime(block['timestamp'], unit='s'),
            'Transaction Hash': tx['hash'].hex(),
            'From': tx['from'],
            'To': tx['to'],
            'Value': web3.fromWei(tx['value'], 'ether'),
            'Nonce': tx['nonce'],
            'Gas': tx['gas'],
            'Gas Price': tx['gasPrice'],
            'Receipt Cumulative': receipt['cumulativeGasUsed']
        })

    return transaction_list

all_transactions = []

# export transactions
for block_number in block_numbers:
    transactions = export_transactions(block_number)
    all_transactions.extend(transactions)
df = pd.DataFrame(all_transactions)

# Save the DataFrame to a CSV file
df.to_csv('transaction_export.csv', index=False)

print(df)

```
